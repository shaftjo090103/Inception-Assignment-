Problem 1: Temperature Check

```python
temp = int(input("What is the current temperature in celcius? "))
if temp < 10 :
  print("It's too cold!")

elif temp >= 10 and temp <= 25 :
  print("It's cool outside")

else :
  print("It's too hot!")
```

Problem 2: ATM Withdrawal System

```python
account_balance = int(input("Enter Your account balance: "))
withdrawal_amount = int(input("Enter the amount you want to withdraw:"))

if account_balance < withdrawal_amount :
  print("Insufficient Balance")

else :
  print("Transaction Successful")
  print("Current balance :",account_balance - withdrawal_amount)

```

Problem 3: Student Grade Calculator

```python
score = int(input("Enter your score :"))

if score >= 90 and score <= 100 :
  print("Grade A+")

elif score >= 80 and score <=89 :
  print("Grade A")

elif score >= 70 and score <=79 :
  print("Grade B")

elif score >= 60 and score <=69 :
  print("Grade C")

elif score > 100 or score < 0 :
  print("Invalid Score")

else :
  print("Fail")
```

Problem 4: Bus Ticket Discount

```python
age = int(input("Enter Your Age :"))
if age < 5 :
  print("Free")

elif age >= 5 and age <= 18 :
  print("50% Discount!")

elif age >= 60 :
  print("30% Discount!")

else :
  print("Full Price")
```

Problem 5: Restaurant Bill with Discount

```python
total_bill = int(input("Enter the total bill amount :"))

if total_bill > 1000 :
  print("You are eligible for 10% discount on your total bill")
  print("Please pay",total_bill - (total_bill * 0.1))

else :
  print("You are not eligible for discount, please pay",total_bill)
```

Problem 7: Traffic Light System

```python
color = input("Enter the color of the traffic light :")

if color == "red" :
  print("Stop")

elif color == "yellow" :
  print("Get Ready")

elif color == "green" :
  print("Go")

else :
  print("Invalid Color")
```

Problem 8: Leap Year Checker

```python
year = int(input("Enter the year :"))

if year % 400 == 0 :
  print("Leap Year")

elif year % 4 == 0 and year % 100 != 0 :
  print("Leap Year")

else :
  print("Not a Leap Year")
```
